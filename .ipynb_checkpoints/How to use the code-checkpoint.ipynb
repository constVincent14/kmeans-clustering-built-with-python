### Import modules and load dataset

```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kmeans_clustering
sns.set_theme(context='notebook', style='darkgrid', palette='flare')
%matplotlib inline

# visualize the data
x_data = np.load('./ex7_X.npy')

fig, axis = plt.subplots(figsize=[9/1.5, 7/1.5])
g = sns.scatterplot(x=x_data[:, 0], y=x_data[:, 1], ax=axis)
g.set(xlim=(0,9), ylim=(0, 7), xlabel='Feature 1', ylabel='Feature 2')
g.set_title("Data Examples")
fig.savefig('./data_examples.jpg', dpi=300)
plt.show()
```

### Build kmeans clustering model
1. First, we need to specify initial centroids, you could create initial centroids and store it in a numpy.ndarray with the shape of `(k, n)`. Where `k` is the number of clusters we want the model to learn and `n` is the number of features or variables in the dataset.
2. One thing to keep in mind when choosing initial centroids is to make sure the total distortion cost not so high. We could manually create it, or using `.fit()` method from the class `kmeans_model`. Hence, when fitting the model, It'will converge much faster and produce a good quality clusters.
3. Other than finding initial centroids with lowest cost, `.fit()` method is also fitting a kmeans algorithm to the dataset.
4. We could use `.f_centroids` attribute to retrieve the final centroids learned by the model. And `.compute_distance()` method to get the clusters for the dataset.

#### Part 1: Automatically finding initial centroids

```python
model = kmeans_clustering.kmeans_model(x_arr=x_data, k=3)
model = model.fit(printout=True, intervals=2)

# get the estimated clusters for the dataset
final_c = model.f_centroids
clusters = model.compute_distance(final_c) 

print(f"Final centroids: \n{final_c}")
print(f"\nDistortion cost at initial: \n{model.score()}")
print(f"\nDistortion cost at final: \n{model.score(model.f_centroids)}")
```

#### Part 2: Manually create initial centroids

```python
# manually guess initial centroids
i_centers = np.array([
    [4, 2],
    [5, 2],
    [5, 3]
])

model = kmeans_clustering.kmeans_model(x_arr=x_data, k=3, random_init_centroids=False, init_centroids=i_centers)
model = model.fit(printout=True, intervals=2)

# get the estimated clusters for the dataset
final_c = model.f_centroids
clusters = model.compute_distance(final_c) 

print(f"Final centroids: \n{final_c}")
print(f"\nDistortion cost at initial: \n{model.score()}")
print(f"\nDistortion cost at final: \n{model.score(model.f_centroids)}")
```

<div class="alert alert-block alert-info">
<b>Note:</b> You could also specify how many iterations for the kmeans model to learn or just turn off the printout.
</div>

```python
# changing num_iters
model = kmeans_clustering.kmeans_model(x_arr=x_data, k=3, random_init_centroids=False, init_centroids=i_centers)
model = model.fit(num_iters=5, printout=True, intervals=None)

print("Model A\nFinal centroids: \n{}".format(model.f_centroids))

# turn off printout
model = kmeans_clustering.kmeans_model(x_arr=x_data, k=3, random_init_centroids=False, init_centroids=i_centers)
model = model.fit(num_iters=7, printout=False, intervals=None)

print("\nModel B\nFinal centroids: \n{}".format(model.f_centroids))
```

### Visualize the result

```python
fig, axis = plt.subplots(figsize=[9/1.5, 7/1.5])
g = sns.scatterplot(x=x_data[:, 0], y=x_data[:, 1], ax=axis, hue=clusters)
g = sns.scatterplot(x=final_c[:, 0], y=final_c[:, 1], ax=axis, label='Centroids')
g.set(xlim=(0,9), ylim=(0, 7), xlabel='Feature 1', ylabel='Feature 2')
g.set_title("Data Examples")
fig.savefig('./clustering_result.jpg', dpi=300)
plt.show()
```
