### Import modules and load dataset

```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kmeans
sns.set_theme(context='notebook', style='darkgrid', palette='flare')
%matplotlib inline

# visualize the data
x_data = np.load('./ex7_X.npy')

fig, axis = plt.subplots(figsize=[9/1.5, 7/1.5])
g = sns.scatterplot(x=x_data[:, 0], y=x_data[:, 1], ax=axis)
g.set(xlim=(0,9), ylim=(0, 7), xlabel='Feature 1', ylabel='Feature 2')
g.set_title("Data Examples")
#fig.savefig('./data_examples.jpg', dpi=300)
plt.show()
```

### Build kmeans clustering algorithm
1. First, we need to specify initial centroids, you could create or guess your own initial centroids and store it in a numpy.ndarray with the shape of `(k, n)`. Where `k` is the number of clusters we want the model to learn and `n` is the number of features or variables in the dataset.
2. One thing to keep in mind is to make sure the total distortion cost (you could use `kmeans.distortion_cost`) of the data is pretty low, for ease I've also provided a function `kmeans.random_init_centroids` to generate randomized initial centroids. The parameters are: `x_arr` **(the dataset)**, `k` **(number of clusters)**, `num_iters` **(number of iterations to search for optimum centroids)**.
3. Once we specify the initial centroids, we then run kmeans on the data. We could use `kmeans.run_kmeans` to do that. The parameters are: `x_arr` **(the dataset)**, `in_centroids` **(initial centroids)**, `num_iters` **(number of iterations)**, `printout` **(print the distortion cost at every `num_iters/5`)**.

```python
# initialize randomized centroids
cost_in, in_centroids = kmeans.random_init_centroids(x_data, k=3, num_iters=100)
fin_centroids, clusters = kmeans.run_kmeans(x_data, in_centroids, 20, printout=True)
```

### Visualize the result

```python
fig, axis = plt.subplots(figsize=[9/1.5, 7/1.5])
g = sns.scatterplot(x=x_data[:, 0], y=x_data[:, 1], ax=axis, hue=clusters)
g = sns.scatterplot(x=fin_centroids[:, 0], y=fin_centroids[:, 1], ax=axis, label='Centroids')
g.set(xlim=(0,9), ylim=(0, 7), xlabel='Feature 1', ylabel='Feature 2')
g.set_title("Data Examples")
#fig.savefig('./clustering_result.jpg', dpi=300)
plt.show()
```
